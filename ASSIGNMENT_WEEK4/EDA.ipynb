```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

df = sns.load_dataset('titanic')

print(df.info())
display(df.describe(include='all'))
```

```python

print(df.isnull().sum())


msno.matrix(df)
plt.show()

msno.heatmap(df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()


sns.boxplot(x=df['age'])
plt.title('Boxplot of Age')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['fare'], bins=40, kde=True, color='salmon')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()


sns.boxplot(x=df['fare'])
plt.title('Boxplot of Fare')
plt.show()
```

```python

categorical_cols = ['sex', 'class', 'embarked', 'who', 'alone']

for col in categorical_cols:
    sns.countplot(data=df, x=col, palette='pastel')
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python

sns.countplot(data=df, x='sex', hue='survived', palette='Set2')
plt.title('Survival Count by Sex')
plt.show()


sns.countplot(data=df, x='class', hue='survived', palette='Set1')
plt.title('Survival Count by Passenger Class')
plt.show()

sns.countplot(data=df, x='embarked', hue='survived', palette='Set3')
plt.title('Survival Count by Embarked Port')
plt.show()
```

```python
sns.violinplot(data=df, x='survived', y='age', palette='muted')
plt.title('Age Distribution by Survival')
plt.show()

sns.boxplot(data=df, x='survived', y='fare', palette='coolwarm')
plt.title('Fare by Survival')
plt.show()
```

```python

num_cols = ['age', 'fare', 'pclass', 'sibsp', 'parch', 'survived']
corr = df[num_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python

df['family_size'] = df['sibsp'] + df['parch'] + 1


sns.boxplot(data=df, x='survived', y='family_size', palette='Blues')
plt.title('Family Size by Survival')
plt.show()
```
